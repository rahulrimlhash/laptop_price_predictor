```python
def processortypedetector(text):
    
    if text=='Intel Core i7' or text=='Intel Core i5' or text=='Intel Core i3':
        return text
    
    else:
        if text.split()[0]=='Intel':
            return 'Other Intel Processor'
        
        else:
            return 'AMD Processor'
        
    
    
data['CPU'] = data['CPU'].apply(lambda text:processortypedetector(text))
data.head()
```

```python
# Using Regular expression to clean the Memory Column
data['Memory'] = data['Memory'].astype(str).replace('\.0','',regex = True)

# replacing the GB word with " "

data['Memory'] = data['Memory'].str.replace('GB','')

# replacing the TB word with "000"

data['Memory'] = data['Memory'].str.replace('TB','000')

# split the word accross the "+" character

new_df = data['Memory'].str.split("+",n = 1,expand = True)

new_df
```

```python
data['First'] = new_df[0]
data['First'] = data['First'].str.strip()
data.head()
```

```python
def Memory_preprocessing(value):
    
    data['Layer1' + value] = data['First'].apply(lambda x:1 if value in x else 0)
    
    
list = ['HDD','SSD','Hybrid','FlashStorage']    
for value in list:
    Memory_preprocessing(value)
      
data.head()
```

```python
data['First'] = data['First'].str.replace(r'\D','')
data['First'].value_counts()
```

```python
data['Second'] = new_df[1]
data.head()
```

```python
def change(value):
    
    data['Layer2'+value] = data['Second'].apply(lambda x:1 if value in x else 0)
    
    
list = ['HDD','SSD','Hybrid','FlashStorage']
data['Second'] = data['Second'].fillna("0")
for value in list:
    change(value)
    

data['Second'] = data['Second'].str.replace(r'\D','')
data['Second'].value_counts()
```

### Machine Learning Model Creation

##### Appying Decision tree algoritms and calculating their accuracy score and finding the best model

```python
test = np.log(data['Price'])
train = data.drop(['Price'],axis = 1)
```

```python
# importing all the modules for Machine Learning
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn import metrics
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
X_train, X_test, y_train, y_test = train_test_split(train, test, test_size=0.15, random_state=2)

X_train.shape, X_test.shape
```

```python
map = {i:value for i,value in enumerate(X_train.columns)}
map
```

## Decision Tree

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,1,3,8,11])
],remainder='passthrough')

step2 = DecisionTreeRegressor(max_depth=8)

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',metrics.r2_score(y_test,y_pred))
print('Mean_Absolute_Error',metrics.mean_absolute_error(y_test,y_pred))
```

## Random Forest Algorithm

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,1,3,8,11])
],remainder='passthrough')

step2 = RandomForestRegressor(n_estimators=100,
                              random_state=3,
                              max_samples=0.5,
                              max_features=0.75,
                              max_depth=15)

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',metrics.r2_score(y_test,y_pred))
print('Mean_Absolute_Error',metrics.mean_absolute_error(y_test,y_pred))
```
